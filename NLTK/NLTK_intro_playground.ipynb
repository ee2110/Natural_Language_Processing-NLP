**Fine-grained Selection**

```python
V = set(text8)
long_words = [w for w in V if len(w) > 12]
sorted(long_words)
```

```python
# select most frequent and long words to eliminate frequent short words, eg. 'a'
fdist8 = FreqDist(text8)
sorted(w for w in set(text8) if len(w) > 7 and fdist8[w] > 10)
```

**Collocations and Bigrams**

A collocation is a sequence of words that occur together unusually often. Thus red wine is a collocation, whereas the wine is not.

To get a handle on collocations, we start off by extracting from a text a list of word pairs, also known as bigrams.

```python
list(bigrams(['more', 'is', 'said', 'than', 'done']))
```

```python
bigrams(['more', 'is', 'said', 'than', 'done'])
```

```python
text8.collocations()
```

```python
#  distribution of word lengths
fdist = FreqDist(len(w) for w in text8)
fdist
```

```python
fdist.most_common()
```

```python
fdist.max()
```

```python
fdist.freq(1)
```

It shows that **words of length 1** account for roughly 30% of the words making up the whole content

```python
fdist.tabulate()
```

```python
fdist.plot()
```

```python
for fileid in gutenberg.fileids():
  num_chars = len(gutenberg.raw(fileid))
  num_words = len(gutenberg.words(fileid))
  num_sents = len(gutenberg.sents(fileid))
  num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
  print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```
