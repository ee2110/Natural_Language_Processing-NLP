**Named-entity recognition (NER)** is the process of extract information and identify the elements in unstructured or structured text by categorize the entities such as person, places, organizations, time, events, etc.

```python
import spacy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ner = spacy.load("en_core_web_sm")
```

Try on some example sentences

```python
text1 = 'Scientists have announced today the stunning discovery of a new species of cat'
doc = ner(text1)
print([(X.text, X.label_) for X in doc.ents])
```

Seems like many general terms were not identified, such as scientists and cat.

```python
text2 = 'I travel to Japan for works and I live at Tokyo for three years'
doc = ner(text2)
print([(X.text, X.label_) for X in doc.ents])
```
